```python
import pandas as pd
df_train = pd.read_csv(r"C:\Users\5kyle\OneDrive\Desktop\titanic\data\train.csv")
df_test = pd.read_csv(r"C:\Users\5kyle\OneDrive\Desktop\titanic\data\test.csv")

df_train.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

survival_by_sex = df_train.groupby("Sex")['Survived'].mean().reset_index()
survival_by_sex
```

```python
sns.barplot(x='Sex', y='Survived', data=survival_by_sex)
plt.ylim(0, 1)
plt.title("Survival by Sex")
plt.ylabel("Proportion Survived")
plt.show()
```
